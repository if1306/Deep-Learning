```python
# 3.1
import numpy as np

print("=" * 60)
print("3.1 理论计算题：贝叶斯公式 - 检测阳性后真正患病的概率")
print("=" * 60)

# 已知条件
prevalence = 0.001  # 患病率 0.1% = 0.001
sensitivity = 0.99   # 灵敏度：患病者检测阳性的概率 99%
false_positive = 0.02  # 假阳性率：未患病者检测阳性的概率 2%

# 计算先验概率
P_disease = prevalence  # P(患病)
P_no_disease = 1 - prevalence  # P(未患病)

# 条件概率
P_positive_given_disease = sensitivity  # P(阳性|患病)
P_positive_given_no_disease = false_positive  # P(阳性|未患病)

# 全概率公式：P(阳性) = P(阳性|患病)P(患病) + P(阳性|未患病)P(未患病)
P_positive = (P_positive_given_disease * P_disease + 
              P_positive_given_no_disease * P_no_disease)

# 贝叶斯公式：P(患病|阳性) = P(阳性|患病)P(患病) / P(阳性)
P_disease_given_positive = (P_positive_given_disease * P_disease) / P_positive

print("\n已知条件：")
print(f"  患病率 P(患病) = {prevalence * 100}% = {prevalence}")
print(f"  灵敏度 P(阳性|患病) = {sensitivity * 100}% = {sensitivity}")
print(f"  假阳性率 P(阳性|未患病) = {false_positive * 100}% = {false_positive}")

print("\n计算过程：")
print(f"  1. P(阳性) = P(阳性|患病)×P(患病) + P(阳性|未患病)×P(未患病)")
print(f"            = {sensitivity} × {prevalence} + {false_positive} × {1 - prevalence:.3f}")
print(f"            = {sensitivity * prevalence:.6f} + {false_positive * (1 - prevalence):.6f}")
print(f"            = {P_positive:.6f}")

print(f"\n  2. P(患病|阳性) = [P(阳性|患病) × P(患病)] / P(阳性)")
print(f"                  = ({sensitivity} × {prevalence}) / {P_positive:.6f}")
print(f"                  = {sensitivity * prevalence:.6f} / {P_positive:.6f}")
print(f"                  = {P_disease_given_positive:.6f}")

print(f"\n结果：")
print(f"  检测结果为阳性的人真正患病的概率为 {P_disease_given_positive * 100:.4f}%")
print(f"  即约 {P_disease_given_positive:.2%}")

# 验证与解释
print("\n" + "=" * 60)
print("解释：")
print(f"  虽然检测灵敏度很高（{sensitivity * 100}%），但由于患病率极低（{prevalence * 100}%），")
print(f"  即使检测呈阳性，真正患病的概率也只有 {P_disease_given_positive * 100:.2f}%。")
print("  这是因为假阳性数量远多于真阳性数量。")

# 可视化（可选）
print("\n" + "=" * 60)
print("数量级示例（假设检测 100 万人）：")
population = 1_000_000
true_disease = int(population * prevalence)
no_disease = population - true_disease

true_positive = int(true_disease * sensitivity)
false_positive_count = int(no_disease * false_positive)
total_positive = true_positive + false_positive_count

print(f"  总人口：{population:,} 人")
print(f"  实际患病：{true_disease:,} 人")
print(f"  实际未患病：{no_disease:,} 人")
print(f"\n  检测阳性人数：")
print(f"    - 真阳性（患病且检测阳性）：{true_positive:,} 人")
print(f"    - 假阳性（未患病但检测阳性）：{false_positive_count:,} 人")
print(f"    - 总阳性：{total_positive:,} 人")
print(f"\n  阳性中真正患病的比例：{true_positive:,} / {total_positive:,} = {true_positive/total_positive:.2%}")

print("=" * 60)
```

```python
# 3.2
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 参数设置
n = 10000  # 每次采样的样本数
m = 1000   # 重复次数

# 生成 m 个均值
means = []
for _ in range(m):
    samples = np.random.uniform(0, 1, n)
    means.append(np.mean(samples))

# 绘制直方图
plt.hist(means, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# 理论正态分布：均值为 0.5，方差为 (1/12)/n
mu_theory = 0.5
sigma_theory = np.sqrt(1/12 / n)
x = np.linspace(min(means), max(means), 100)
plt.plot(x, norm.pdf(x, mu_theory, sigma_theory), 'r-', label='Theoretical Normal')

plt.title("Central Limit Theorem Simulation")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()

# 实际方差
actual_var = np.var(means)
print("Actual variance of sample means:", actual_var)
```

```python
# 4.1
import numpy as np

print("=" * 60)
print("4.1 理论计算题：梯度计算")
print("=" * 60)

# 给定常数
x1 = 2
x2 = 1
y = 3

# 函数：z = (w1*x1 + w2*x2 - y)^2

print("\n已知条件：")
print(f"  x1 = {x1}, x2 = {x2}, y = {y}")
print(f"  z = (w1×{x1} + w2×{x2} - {y})² = ({x1}w1 + {x2}w2 - {y})²")

print("\n" + "=" * 60)
print("1. 偏导数表达式推导")
print("=" * 60)

print("\n令 u = w1*x1 + w2*x2 - y")
print("则 z = u²")

print("\n根据链式法则：")
print("  ∂z/∂w1 = ∂z/∂u × ∂u/∂w1")
print("  ∂z/∂w2 = ∂z/∂u × ∂u/∂w2")

print("\n计算各项：")
print("  ∂z/∂u = 2u")
print("  ∂u/∂w1 = x1 =", x1)
print("  ∂u/∂w2 = x2 =", x2)

print("\n因此：")
print(f"  ∂z/∂w1 = 2u × x1 = 2 × ({x1}w1 + {x2}w2 - {y}) × {x1}")
print(f"         = {2*x1} × ({x1}w1 + {x2}w2 - {y})")
print(f"         = {2*x1*x1}w1 + {2*x1*x2}w2 - {2*x1*y}")

print(f"\n  ∂z/∂w2 = 2u × x2 = 2 × ({x1}w1 + {x2}w2 - {y}) × {x2}")
print(f"         = {2*x2} × ({x1}w1 + {x2}w2 - {y})")
print(f"         = {2*x2*x1}w1 + {2*x2*x2}w2 - {2*x2*y}")

print("\n" + "=" * 60)
print("2. 代入具体数值计算梯度")
print("=" * 60)

# 给定的参数值
w1 = 0.5
w2 = 1

print(f"\n当前参数：")
print(f"  w1 = {w1}, w2 = {w2}")

# 计算 u
u = w1 * x1 + w2 * x2 - y
print(f"\n中间变量 u = {w1}×{x1} + {w2}×{x2} - {y}")
print(f"        = {w1*x1} + {w2*x2} - {y}")
print(f"        = {u}")

# 计算 z
z = u ** 2
print(f"\n函数值 z = u² = {u}² = {z}")

# 计算梯度（解析解）
dz_dw1 = 2 * u * x1
dz_dw2 = 2 * u * x2

print(f"\n梯度计算（解析解）：")
print(f"  ∂z/∂w1 = 2 × u × x1 = 2 × {u} × {x1} = {dz_dw1}")
print(f"  ∂z/∂w2 = 2 × u × x2 = 2 × {u} × {x2} = {dz_dw2}")

print(f"\n梯度向量：∇z = [{dz_dw1}, {dz_dw2}]")

# 数值验证（使用有限差分法）
print("\n" + "=" * 60)
print("3. 数值验证（有限差分法）")
print("=" * 60)

epsilon = 1e-6

# 对 w1 进行数值微分
def compute_z(w1_val, w2_val):
    u_val = w1_val * x1 + w2_val * x2 - y
    return u_val ** 2

# 数值梯度 d/dw1
z_plus = compute_z(w1 + epsilon, w2)
z_minus = compute_z(w1 - epsilon, w2)
numerical_dw1 = (z_plus - z_minus) / (2 * epsilon)

# 数值梯度 d/dw2
z_plus = compute_z(w1, w2 + epsilon)
z_minus = compute_z(w1, w2 - epsilon)
numerical_dw2 = (z_plus - z_minus) / (2 * epsilon)

print(f"\n数值梯度结果：")
print(f"  数值 ∂z/∂w1 = {numerical_dw1:.10f}")
print(f"  数值 ∂z/∂w2 = {numerical_dw2:.10f}")

print(f"\n解析梯度结果：")
print(f"  解析 ∂z/∂w1 = {dz_dw1:.10f}")
print(f"  解析 ∂z/∂w2 = {dz_dw2:.10f}")

# 验证
tolerance = 1e-8
dw1_correct = abs(dz_dw1 - numerical_dw1) < tolerance
dw2_correct = abs(dz_dw2 - numerical_dw2) < tolerance

print(f"\n验证结果：")
print(f"  ∂z/∂w1 正确: {dw1_correct} (差异: {abs(dz_dw1 - numerical_dw1):.2e})")
print(f"  ∂z/∂w2 正确: {dw2_correct} (差异: {abs(dz_dw2 - numerical_dw2):.2e})")

print("\n" + "=" * 60)
print("最终答案：")
print("=" * 60)
print(f"1. 偏导数表达式：")
print(f"   ∂z/∂w1 = {2*x1}·({x1}w1 + {x2}w2 - {y}) = {2*x1*x1}w1 + {2*x1*x2}w2 - {2*x1*y}")
print(f"   ∂z/∂w2 = {2*x2}·({x1}w1 + {x2}w2 - {y}) = {2*x2*x1}w1 + {2*x2*x2}w2 - {2*x2*y}")
print(f"\n2. 当 w1={w1}, w2={w2} 时：")
print(f"   ∂z/∂w1 = {dz_dw1}")
print(f"   ∂z/∂w2 = {dz_dw2}")
print("=" * 60)
```

```python
# 4.2
import numpy as np

# 定义前向计算函数
def forward(x, w1, w2):
    a = x * w1
    b = a + w2
    L = b ** 2
    return L, a, b

# 定义反向传播函数（手动计算梯度）
def backward(x, w1, w2, a, b):
    # 链式法则
    dL_db = 2 * b          # dL/db = 2b
    db_da = 1              # db/da = 1
    db_dw2 = 1             # db/dw2 = 1
    da_dw1 = x             # da/dw1 = x
    
    dL_dw1 = dL_db * db_da * da_dw1
    dL_dw2 = dL_db * db_dw2
    
    return dL_dw1, dL_dw2

# 给定参数
x = 2.0
w1 = 1.5
w2 = 0.5

# 前向计算
L, a, b = forward(x, w1, w2)
print(f"前向计算结果:")
print(f"a = x * w1 = {x} * {w1} = {a}")
print(f"b = a + w2 = {a} + {w2} = {b}")
print(f"L = b^2 = {b}^2 = {L}")
print()

# 手动计算梯度
dL_dw1, dL_dw2 = backward(x, w1, w2, a, b)
print(f"手动计算的梯度:")
print(f"dL/dw1 = {dL_dw1}")
print(f"dL/dw2 = {dL_dw2}")
print()

# 数值验证（使用有限差分法，验证手动梯度是否正确）
epsilon = 1e-5

# 数值梯度 dL/dw1
L1, _, _ = forward(x, w1 + epsilon, w2)
L2, _, _ = forward(x, w1 - epsilon, w2)
numerical_dw1 = (L1 - L2) / (2 * epsilon)

# 数值梯度 dL/dw2
L1, _, _ = forward(x, w1, w2 + epsilon)
L2, _, _ = forward(x, w1, w2 - epsilon)
numerical_dw2 = (L1 - L2) / (2 * epsilon)

print(f"数值验证的梯度（有限差分法）:")
print(f"数值 dL/dw1 = {numerical_dw1}")
print(f"数值 dL/dw2 = {numerical_dw2}")
print()

# 验证手动梯度是否正确
tolerance = 1e-6
dw1_correct = abs(dL_dw1 - numerical_dw1) < tolerance
dw2_correct = abs(dL_dw2 - numerical_dw2) < tolerance

print(f"手动梯度验证结果:")
print(f"dL/dw1 正确: {dw1_correct}")
print(f"dL/dw2 正确: {dw2_correct}")
```

```python
# 5.1
import numpy as np
import matplotlib.pyplot as plt

print("=" * 70)
print("5.1 理论计算题：线性回归梯度推导")
print("=" * 70)

# 理论推导部分（打印）
print("\n" + "=" * 70)
print("1. 理论推导")
print("=" * 70)

print("\n均方误差损失函数：")
print("  L = (1/n) · Σᵢ₌₁ⁿ (yᵢ - (w xᵢ + b))²")

print("\n令残差 rᵢ = yᵢ - (w xᵢ + b)，则 L = (1/n) · Σᵢ₌₁ⁿ rᵢ²")

print("\n对 w 求偏导：")
print("  ∂L/∂w = (1/n) · Σᵢ₌₁ⁿ 2·rᵢ · ∂rᵢ/∂w")
print("  ∂rᵢ/∂w = ∂/∂w [yᵢ - (w xᵢ + b)] = -xᵢ")
print("  所以 ∂L/∂w = (1/n) · Σᵢ₌₁ⁿ 2·rᵢ · (-xᵢ)")
print("          = -(2/n) · Σᵢ₌₁ⁿ xᵢ·(yᵢ - (w xᵢ + b))")

print("\n对 b 求偏导：")
print("  ∂L/∂b = (1/n) · Σᵢ₌₁ⁿ 2·rᵢ · ∂rᵢ/∂b")
print("  ∂rᵢ/∂b = ∂/∂b [yᵢ - (w xᵢ + b)] = -1")
print("  所以 ∂L/∂b = (1/n) · Σᵢ₌₁ⁿ 2·rᵢ · (-1)")
print("          = -(2/n) · Σᵢ₌₁ⁿ (yᵢ - (w xᵢ + b))")

print("\n最终表达式：")
print("  ∂L/∂w = -(2/n) · Σᵢ₌₁ⁿ xᵢ·(yᵢ - (w xᵢ + b))")
print("  ∂L/∂b = -(2/n) · Σᵢ₌₁ⁿ (yᵢ - (w xᵢ + b))")

print("\n" + "=" * 70)
print("2. 数值验证（使用示例数据）")
print("=" * 70)

# 生成示例数据
np.random.seed(42)
n = 100
X = np.linspace(0, 10, n)
true_w = 2.5
true_b = 1.0
y = true_w * X + true_b + np.random.randn(n) * 2  # 添加噪声

# 初始化参数
w = 1.0
b = 0.0
learning_rate = 0.01

print(f"\n生成 {n} 个样本点")
print(f"真实模型: y = {true_w}·x + {true_b}")
print(f"初始参数: w = {w}, b = {b}")

# 计算梯度（使用推导的公式）
def compute_gradients(X, y, w, b):
    n = len(X)
    y_pred = w * X + b
    residuals = y - y_pred
    
    grad_w = -(2/n) * np.sum(X * residuals)
    grad_b = -(2/n) * np.sum(residuals)
    
    return grad_w, grad_b

# 计算初始梯度
grad_w, grad_b = compute_gradients(X, y, w, b)

print(f"\n初始梯度：")
print(f"  ∂L/∂w = {grad_w:.4f}")
print(f"  ∂L/∂b = {grad_b:.4f}")

# 数值验证（使用有限差分法）
print("\n" + "=" * 70)
print("3. 数值验证（有限差分法）")
print("=" * 70)

def compute_loss(X, y, w, b):
    y_pred = w * X + b
    return np.mean((y - y_pred) ** 2)

epsilon = 1e-6

# 数值梯度 dL/dw
loss_plus = compute_loss(X, y, w + epsilon, b)
loss_minus = compute_loss(X, y, w - epsilon, b)
numerical_grad_w = (loss_plus - loss_minus) / (2 * epsilon)

# 数值梯度 dL/db
loss_plus = compute_loss(X, y, w, b + epsilon)
loss_minus = compute_loss(X, y, w, b - epsilon)
numerical_grad_b = (loss_plus - loss_minus) / (2 * epsilon)

print(f"\n数值梯度结果：")
print(f"  数值 ∂L/∂w = {numerical_grad_w:.10f}")
print(f"  数值 ∂L/∂b = {numerical_grad_b:.10f}")

print(f"\n解析梯度结果：")
print(f"  解析 ∂L/∂w = {grad_w:.10f}")
print(f"  解析 ∂L/∂b = {grad_b:.10f}")

# 验证
tolerance = 1e-8
dw_correct = abs(grad_w - numerical_grad_w) < tolerance
db_correct = abs(grad_b - numerical_grad_b) < tolerance

print(f"\n验证结果：")
print(f"  ∂L/∂w 正确: {dw_correct} (差异: {abs(grad_w - numerical_grad_w):.2e})")
print(f"  ∂L/∂b 正确: {db_correct} (差异: {abs(grad_b - numerical_grad_b):.2e})")

# 演示梯度下降
print("\n" + "=" * 70)
print("4. 梯度下降演示（使用推导的梯度）")
print("=" * 70)

w_hist = [w]
b_hist = [b]
loss_hist = [compute_loss(X, y, w, b)]

for epoch in range(100):
    grad_w, grad_b = compute_gradients(X, y, w, b)
    w -= learning_rate * grad_w
    b -= learning_rate * grad_b
    
    w_hist.append(w)
    b_hist.append(b)
    loss_hist.append(compute_loss(X, y, w, b))

print(f"\n训练100次迭代后：")
print(f"  最终 w = {w:.4f} (真实值: {true_w})")
print(f"  最终 b = {b:.4f} (真实值: {true_b})")
print(f"  最终损失 = {loss_hist[-1]:.4f}")

# 可视化
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# 左图：数据拟合结果
axes[0].scatter(X, y, alpha=0.5, label='真实数据', s=10)
X_line = np.array([0, 10])
y_initial = 1.0 * X_line + 0.0
y_final = w * X_line + b
y_true_line = true_w * X_line + true_b

axes[0].plot(X_line, y_initial, 'g--', label='初始模型', linewidth=2)
axes[0].plot(X_line, y_final, 'r-', label='训练后模型', linewidth=2)
axes[0].plot(X_line, y_true_line, 'b:', label='真实模型', linewidth=2)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('线性回归拟合结果')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# 右图：损失下降曲线
axes[1].plot(loss_hist, 'b-', linewidth=2)
axes[1].set_xlabel('迭代次数')
axes[1].set_ylabel('损失值')
axes[1].set_title('损失函数下降曲线')
axes[1].set_yscale('log')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "=" * 70)
print("最终答案：")
print("=" * 70)
print("∂L/∂w = -(2/n) · Σᵢ₌₁ⁿ xᵢ·(yᵢ - (w xᵢ + b))")
print("∂L/∂b = -(2/n) · Σᵢ₌₁ⁿ (yᵢ - (w xᵢ + b))")
print("\n其中 n 是样本数量")
print("=" * 70)
```

```python
# 5.2
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

# 加载数据
digits = load_digits()
X = digits.data  # (1797, 64)
y = digits.target.reshape(-1, 1)

# One-hot 编码
encoder = OneHotEncoder(sparse_output=False)
y_onehot = encoder.fit_transform(y)

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y_onehot, test_size=0.2, random_state=42)

# Softmax 函数
def softmax(z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

# 交叉熵损失
def cross_entropy_loss(y_pred, y_true):
    return -np.mean(np.sum(y_true * np.log(y_pred + 1e-8), axis=1))

# 参数初始化
input_dim = X_train.shape[1]
num_classes = y_train.shape[1]
W = np.random.randn(input_dim, num_classes) * 0.01
b = np.zeros((1, num_classes))

# 小批量 SGD
lr = 0.1
batch_size = 32
epochs = 50

for epoch in range(epochs):
    # 打乱数据
    indices = np.random.permutation(len(X_train))
    X_shuffled = X_train[indices]
    y_shuffled = y_train[indices]

    for i in range(0, len(X_train), batch_size):
        X_batch = X_shuffled[i:i+batch_size]
        y_batch = y_shuffled[i:i+batch_size]

        # 前向传播
        logits = X_batch @ W + b
        y_pred = softmax(logits)

        # 梯度计算
        grad_logits = y_pred - y_batch
        grad_W = X_batch.T @ grad_logits / batch_size
        grad_b = np.mean(grad_logits, axis=0, keepdims=True)

        # 更新参数
        W -= lr * grad_W
        b -= lr * grad_b

    # 打印损失
    if epoch % 10 == 0:
        logits_train = X_train @ W + b
        y_pred_train = softmax(logits_train)
        loss = cross_entropy_loss(y_pred_train, y_train)
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

# 测试准确率
logits_test = X_test @ W + b
y_pred_test = softmax(logits_test)
y_pred_labels = np.argmax(y_pred_test, axis=1)
y_true_labels = np.argmax(y_test, axis=1)
accuracy = np.mean(y_pred_labels == y_true_labels)
print(f"Test accuracy: {accuracy:.4f}")
```

```python
# 6.1
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

print("=" * 70)
print("6.1 理论计算题：正态分布的最大似然估计")
print("=" * 70)

# ============================================
# 1. 理论推导（打印）
# ============================================
print("\n" + "=" * 70)
print("1. 理论推导")
print("=" * 70)

print("\n似然函数：")
print("  L(μ, σ²) = ∏ᵢ₌₁ⁿ f(xᵢ; μ, σ²)")
print("           = ∏ᵢ₌₁ⁿ [1/√(2πσ²) · exp(-(xᵢ-μ)²/(2σ²))]")
print("           = (2πσ²)⁻ⁿ/² · exp(-∑ᵢ₌₁ⁿ (xᵢ-μ)²/(2σ²))")

print("\n对数似然函数：")
print("  ℓ(μ, σ²) = ln L(μ, σ²)")
print("           = -n/2 · ln(2π) - n/2 · ln(σ²) - 1/(2σ²) · ∑ᵢ₌₁ⁿ (xᵢ-μ)²")

print("\n对 μ 求偏导并令为 0：")
print("  ∂ℓ/∂μ = -1/(2σ²) · ∑ᵢ₌₁ⁿ 2(xᵢ-μ)·(-1) = (1/σ²) · ∑ᵢ₌₁ⁿ (xᵢ-μ) = 0")
print("  ⇒ ∑ᵢ₌₁ⁿ (xᵢ-μ) = 0")
print("  ⇒ ∑ᵢ₌₁ⁿ xᵢ - nμ = 0")
print("  ⇒ μ̂ = (1/n) · ∑ᵢ₌₁ⁿ xᵢ")

print("\n对 σ² 求偏导并令为 0：")
print("  令 θ = σ²")
print("  ∂ℓ/∂θ = -n/(2θ) + 1/(2θ²) · ∑ᵢ₌₁ⁿ (xᵢ-μ)² = 0")
print("  ⇒ -nθ + ∑ᵢ₌₁ⁿ (xᵢ-μ)² = 0")
print("  ⇒ θ̂ = (1/n) · ∑ᵢ₌₁ⁿ (xᵢ-μ)²")
print("  ⇒ σ̂² = (1/n) · ∑ᵢ₌₁ⁿ (xᵢ-μ̂)²")

print("\n最终结果：")
print("  μ̂ = (1/n) · Σᵢ₌₁ⁿ xᵢ")
print("  σ̂² = (1/n) · Σᵢ₌₁ⁿ (xᵢ - μ̂)²")

# ============================================
# 2. 数值验证
# ============================================
print("\n" + "=" * 70)
print("2. 数值验证")
print("=" * 70)

# 生成服从正态分布的数据
np.random.seed(42)
true_mu = 5.0
true_sigma2 = 4.0  # 方差为4，标准差为2
n = 1000

samples = np.random.normal(true_mu, np.sqrt(true_sigma2), n)

print(f"\n生成 {n} 个服从 N({true_mu}, {true_sigma2}) 的样本")
print(f"真实参数: μ = {true_mu}, σ² = {true_sigma2}")

# 计算 MLE 估计
mu_mle = np.mean(samples)
sigma2_mle = np.var(samples)  # 注意：np.var 默认使用 n，即 MLE 估计

print(f"\nMLE 估计结果：")
print(f"  μ̂ = {mu_mle:.6f}")
print(f"  σ̂² = {sigma2_mle:.6f}")

print(f"\n估计误差：")
print(f"  |μ - μ̂| = {abs(true_mu - mu_mle):.6f}")
print(f"  |σ² - σ̂²| = {abs(true_sigma2 - sigma2_mle):.6f}")

# ============================================
# 3. 与样本方差比较（无偏估计）
# ============================================
print("\n" + "=" * 70)
print("3. 与无偏样本方差比较")
print("=" * 70)

sample_var_unbiased = np.var(samples, ddof=1)  # 无偏估计，除以 n-1

print(f"\nMLE 估计（除以 n）: σ̂² = {sigma2_mle:.6f}")
print(f"无偏估计（除以 n-1）: s² = {sample_var_unbiased:.6f}")
print(f"真实方差: {true_sigma2}")

print(f"\n偏差分析：")
print(f"  MLE 偏差 = {sigma2_mle - true_sigma2:.6f}")
print(f"  无偏估计偏差 = {sample_var_unbiased - true_sigma2:.6f}")

# ============================================
# 4. 可视化
# ============================================
print("\n" + "=" * 70)
print("4. 可视化")
print("=" * 70)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# 左图：数据直方图和拟合的正态分布
axes[0].hist(samples, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black', label='样本直方图')

x = np.linspace(min(samples), max(samples), 100)
true_pdf = stats.norm.pdf(x, true_mu, np.sqrt(true_sigma2))
mle_pdf = stats.norm.pdf(x, mu_mle, np.sqrt(sigma2_mle))

axes[0].plot(x, true_pdf, 'r-', linewidth=2, label=f'真实分布 N({true_mu}, {true_sigma2})')
axes[0].plot(x, mle_pdf, 'g--', linewidth=2, label=f'MLE 估计 N({mu_mle:.2f}, {sigma2_mle:.2f})')
axes[0].set_xlabel('x')
axes[0].set_ylabel('密度')
axes[0].set_title('正态分布 MLE 拟合')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# 右图：MLE 估计的收敛性（随着样本量增加）
n_range = np.arange(10, 1001, 10)
mu_estimates = []
sigma2_estimates = []

for n_ in n_range:
    samples_ = np.random.normal(true_mu, np.sqrt(true_sigma2), n_)
    mu_estimates.append(np.mean(samples_))
    sigma2_estimates.append(np.var(samples_))

axes[1].axhline(y=true_mu, color='r', linestyle='-', label=f'真实 μ = {true_mu}')
axes[1].axhline(y=true_sigma2, color='b', linestyle='-', label=f'真实 σ² = {true_sigma2}')
axes[1].plot(n_range, mu_estimates, 'r--', alpha=0.7, label='μ 的 MLE')
axes[1].plot(n_range, sigma2_estimates, 'b--', alpha=0.7, label='σ² 的 MLE')
axes[1].set_xlabel('样本数量 n')
axes[1].set_ylabel('估计值')
axes[1].set_title('MLE 估计的收敛性')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================
# 5. 大样本性质验证
# ============================================
print("\n" + "=" * 70)
print("5. 大样本性质验证（重复实验）")
print("=" * 70)

n_experiments = 1000
n_samples = 100

mu_estimates_many = []
sigma2_estimates_many = []

for _ in range(n_experiments):
    samples_ = np.random.normal(true_mu, np.sqrt(true_sigma2), n_samples)
    mu_estimates_many.append(np.mean(samples_))
    sigma2_estimates_many.append(np.var(samples_))

mu_mean = np.mean(mu_estimates_many)
mu_var = np.var(mu_estimates_many)
sigma2_mean = np.mean(sigma2_estimates_many)
sigma2_var = np.var(sigma2_estimates_many)

print(f"\n重复 {n_experiments} 次实验，每次使用 {n_samples} 个样本：")
print(f"\nμ 的 MLE：")
print(f"  估计均值 = {mu_mean:.6f} (理论值: {true_mu})")
print(f"  估计方差 = {mu_var:.6f} (理论方差: σ²/n = {true_sigma2/n_samples:.6f})")
print(f"\nσ² 的 MLE：")
print(f"  估计均值 = {sigma2_mean:.6f} (理论值: {true_sigma2})")
print(f"  估计方差 = {sigma2_var:.6f}")

# ============================================
# 6. 最终答案
# ============================================
print("\n" + "=" * 70)
print("最终答案")
print("=" * 70)
print("1. 似然函数：")
print("   L(μ, σ²) = (2πσ²)⁻ⁿ/² · exp(-∑ᵢ₌₁ⁿ (xᵢ-μ)²/(2σ²))")
print("\n2. μ 的 MLE：")
print("   μ̂ = (1/n) · Σᵢ₌₁ⁿ xᵢ")
print("\n3. σ² 的 MLE：")
print("   σ̂² = (1/n) · Σᵢ₌₁ⁿ (xᵢ - μ̂)²")
print("=" * 70)
```

```python
# 6.2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# 生成数据
X, y = make_classification(n_samples=400, n_features=2, n_redundant=0, n_clusters_per_class=1, random_state=42)
y = y.reshape(-1, 1)

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=100, random_state=42)

# Sigmoid 函数
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# 二元交叉熵损失
def binary_cross_entropy(y_pred, y_true):
    return -np.mean(y_true * np.log(y_pred + 1e-8) + (1 - y_true) * np.log(1 - y_pred + 1e-8))

# 初始化参数
w = np.zeros((2, 1))
b = 0.0
lr = 0.1
iterations = 1000

for i in range(iterations):
    # 前向传播
    logits = X_train @ w + b
    y_pred = sigmoid(logits)

    # 梯度计算
    grad_w = (X_train.T @ (y_pred - y_train)) / len(X_train)
    grad_b = np.mean(y_pred - y_train)

    # 更新参数
    w -= lr * grad_w
    b -= lr * grad_b

    if i % 200 == 0:
        loss = binary_cross_entropy(y_pred, y_train)
        print(f"Iter {i}, Loss: {loss:.4f}")

# 测试准确率
logits_test = X_test @ w + b
y_pred_test = sigmoid(logits_test) >= 0.5
accuracy = np.mean(y_pred_test == y_test)
print(f"Test accuracy: {accuracy:.4f}")

# 绘制决策边界
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
Z = sigmoid(np.c_[xx.ravel(), yy.ravel()] @ w + b)
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, levels=[0, 0.5, 1], alpha=0.3, colors=['blue', 'red'])
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test.ravel(), edgecolors='k', cmap='bwr')
plt.title("Logistic Regression Decision Boundary")
plt.show()
```
